## CCTV와 인구 현황 데이터 파악하기

```python
CCTV_Seoul.rename( columns = { CCTV_Seoul.columns[1] : '소계'}, inplace = True )
CCTV_Seoul.drop( index = 0 , inplace = True) # 이걸로 계를 삭제 시킴 # 나중에 더 깔끔한 조건 찾아보자 #ToDo CCTV.Seoul.drop ( if 구별 == 계)
CCTV_Seoul.head()
```

```python
CCTV_Seoul.sort_values( by='소계', ascending = False ).head()
```

```python
# 최근 증가율 계산해서 컬럼 추가하기
CCTV_Seoul['최근증가율'] =  (CCTV_Seoul['2018년'] + CCTV_Seoul['2019년'] + CCTV_Seoul['2020년']) / CCTV_Seoul['2018년 이전'] * 100 
CCTV_Seoul.sort_values ( by = '최근증가율', ascending = False ).head()
```

```python
pop_Seoul.head()
```

```python
pop_Seoul.drop( [0], inplace = True) # 합계는 필요없어서 지우기
pop_Seoul.head()
```

```python
pop_Seoul['구별'].unique() # 반복된 데이터는 하나로 나타냄 #반복된 데이터가 있으면 NaN 으로 나옴
```

```python
#외국인비율과 고령자비율 계산
pop_Seoul['외국인비율'] = pop_Seoul['외국인'] / pop_Seoul['인구수'] * 100
pop_Seoul['고령자비율'] = pop_Seoul['고령자'] / pop_Seoul['인구수'] * 100
pop_Seoul.head()
```

## 인구 데이터와 CCTV 데이터 합치고 분석하기

```python
data_result = pd.merge( CCTV_Seoul, pop_Seoul, on='구별') # 공통된 컬럼인 '구별'로 merge
data_result.head()
```

```python
#의미 없는 칼럼 지우기 #행방향 삭제는 drop #열방향 삭제는 del
del data_result['2018년 이전']
del data_result['2018년']
del data_result['2019년']
del data_result['2020년']
data_result.head()
```

```python
# 구 이름을 index 로 변경하기
data_result.set_index('구별', inplace = True)
data_result.head()
```

```python
# 인구 대비 CCTV 비율 계산 후 그래프 확인
data_result['CCTV비율'] = data_result['소계'] / data_result['인구수'] * 100
data_result.head()
```

```python
data_result['CCTV비율'].sort_values().plot(kind='barh', grid=True, figsize=(10, 10))
plt.show()
```

```python
plt.figure(figsize=(6, 6))
plt.scatter(data_result['인구수'], data_result['소계'], s=50) 
plt.xlabel('인구수')
plt.ylabel('CCTV')
plt.grid()
plt.show()
```

```python
fp1 = np.polyfit( data_result['인구수'], data_result['소계'], 1)
fp1
```

```python
f1 = np.poly1d(fp1)
fx = np.linspace(100000, 700000, 100) # 이 데이터들의 대표하는 값들의 직선을 찾는 과정임.
```

```python
plt.figure(figsize=(10, 10))
plt.scatter(data_result['인구수'], data_result['소계'], s=50)
plt.plot(fx, f1(fx), ls='dashed', lw=3, color='g')
plt.xlabel('인구수')
plt.xlabel('CCTV')
plt.grid()
plt.show()
#아래 직선은 인구수에 따른 CCTV의 예상 갯수를 말함.
```

```python
fp1 = np.polyfit( data_result['인구수'], data_result['소계'], 1)
f1 = np.poly1d(fp1)
fx = np.linspace(100000, 700000, 100)

data_result['오차'] = np.abs(data_result['소계'] - f1(data_result['인구수']))

df_sort = data_result.sort_values( by='오차', ascending = False )
df_sort.head()
```

```python
plt.figure(figsize=(14, 10))
plt.scatter(data_result['인구수'], data_result['소계'], c = data_result['오차'], s = 50 ) # 오차에 따라 다른 색깔로 표시
plt.plot(fx, f1(fx), ls='dashed', lw=3, color='g')

for n in range(10):
    plt.text(df_sort['인구수'][n]*1.02, df_sort['소계'][n]*0.98, df_sort.index[n], fontsize=15)
    
plt.xlabel('인구수')
plt.ylabel('인구당비율')

plt.colorbar()
plt.grid()
plt.show()
```

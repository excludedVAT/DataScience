### 두 DataFrame 병합하기

```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'], 
                    'B': ['B0', 'B1', 'B2', 'B3'],
                    'C': ['C0', 'C1', 'C2', 'C3'],
                    'D': ['D0', 'D1', 'D2', 'D3']},
                   index=[0, 1, 2, 3])

df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                    'B': ['B4', 'B5', 'B6', 'B7'],
                    'C': ['C4', 'C5', 'C6', 'C7'],
                    'D': ['D4', 'D5', 'D6', 'D7']},
                   index=[4, 5, 6, 7])

df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                    'B': ['B8', 'B9', 'B10', 'B11'],
                    'C': ['C8', 'C9', 'C10', 'C11'],
                    'D': ['D8', 'D9', 'D10', 'D11']},
                   index=[8, 9, 10, 11])
```

```python
df1
```

```python
df2
```

```python
df3
```

```python
#열 방향으로 단순히 합침
result = pd.concat( [df1, df2, df3])
result
```

```python
#옵션이 없으면 열 방향으로 병합함, 데이터를 아래로 추가하는 개념인듯
#추가된 key는 다중 index가 되어서 level을 형성한다.
result = pd.concat( [df1, df2, df3], keys = ['x', 'y', 'z'])
result
```

```python
result.index
```

```python
result.index.get_level_values(0)
```

```python
result.index.get_level_values(1)
```

```python
df4 = pd.DataFrame({'B': ['B2', 'B3', 'B6', 'B7'], 
                    'D': ['D2', 'D3', 'D6', 'D7'],
                    'F': ['F2', 'F3', 'F6', 'F7']},
                   index=[2, 3, 6, 7])
df4
```

```python
df1
```

```python
result = pd.concat( [df1, df4] ) # axis 옵션이 없을때는 행을 추가
result
```

```python
result = pd.concat( [df1, df4], axis = 1 ) # concat명령은 index를 기준으로 데이터를 합치기 때문에 값을 가지지 않는 곳에서는 NaN가 저장됨
# axis = 1 옵션을 주었을 때는 열을 추가하는 듯
result
```

```python
result = pd.concat( [df1, df4], axis = 1, join='inner' ) # join='inner' 옵션은 공통된 index 가 없을때는 데이터를 버린다.
result
```

```python
result = pd.concat( [df1, df4], axis = 1 ).reindex(df1.index) # df1의 인덱스에 맞추기
result
```

```python
result = pd.concat( [df1, df4], ignore_index = True ) # 열 방향으로 합치면서 기존의 index는 무시하고 새로 부여한다.
result
```

```python
left = pd.DataFrame({'key': ['K0', 'K4', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})

right = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                      'C': ['C0', 'C1', 'C2', 'C3'],
                      'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
left
```

```python
right
```

```python
pd.merge(left, right)
```

```python
pd.merge(left, right, on='key') # key라는 공통 칼럼이 있을때만 합친다.
```

```python
pd.merge(left, right, how='left', on='key') # left를 기준으로 합친다.
```

```python
pd.merge(left, right, how='outer', on='key') # merge 한 데이터 결과를 모두 가지는 outer 옵션, 합집합처럼 머지가 됨
```

```python
pd.merge(left, right, how='inner', on='key') # inner는 반대로 교집합만 머지가 됨
```
